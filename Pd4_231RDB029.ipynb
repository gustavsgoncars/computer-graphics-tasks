```python
### Python pakotņu importēšana
#darbības ar masīviem
import numpy as np
from numpy import asarray

#attēlu apstrādes bibliotēkas
import cv2
from PIL import Image

#attēlu atvēršana/parādīšana Notebook
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

### Attēla izmēra maiņa, mērogošana

image = cv2.imread('Lenna.png')
im_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

resized_img = cv2.resize(im_rgb, (500,500))

#x un y robežas līdz 500
for x in range(500):
    for y in range(500):
        #katra otrā vērtība x un katra trešā vērtība y asī
        if (x % 6==0) or(y % 10==0):
            resized_img[y, x] = (255,255,255)

#%matplotlib notebook
img = Image.fromarray(resized_img)
img.save('Lenna2.png')

plt.imshow(resized_img)
```
